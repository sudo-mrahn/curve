```python
# the gamma distribution
n, shape, scale = int(2e4), 2.05, 10
gamma = genscore.standard(n, shape, scale)
```

```python
# bimodal distribution
# as sum of two normal distributions 
n, p, mod_1, mod_2, sd_1, sd_2 = 2e4, 0.7, 90, 70, 5, 12
bimod = genscore.bimodal(n,p,mod_1,mod_2,sd_1,sd_2)
```

```python
# trimodal distribution
# Most similar to P5 final exam:
n,p_1,p_2,mod_1,mod_2,mod_3,sd_1,sd_2,sd_3 = 2e2, 0.33, 0.4, 92, 80, 60, 2, 6, 12
trimod = genscore.trimodal(n,p_1,p_2,mod_1,mod_2,mod_3,sd_1,sd_2,sd_3)
```

```python
show.compare4(
    trimod, 
    cv.floor(trimod, 50), 
    cv.scalemax(trimod, 50), 
    cv.scalemed(trimod, 50))
```
